```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pymc3 as pm
import seaborn as sns
import arviz as az
```

```python
nba = pd.read_csv("bask_clean.csv", dtype={'Abb':'category'})
```

```python
nba.head()
# Final = total wins out of 82 games 
# Wins = total wins as of Dec 5
# Loss = total losses as of Dec 5
# Games = total games as of Dec 5
# Proj = for the whole year
```

```python
team_idx = nba.Abb.cat.codes.values
team_codes = nba.Abb.cat.categories
n_teams = team_codes.size

coords = {'team': ['ATL', 'BOS', 'BRK', 'CHI', 'CHO', 'CLE', 'DAL', 'DEN', 'DET', 'GSW',
                    'HOU', 'IND', 'LAC', 'LAL', 'MEM', 'MIA', 'MIL', 'MIN', 'NOP', 'NYK',
                    'OKC', 'ORL', 'PHI', 'PHO', 'POR', 'SAC', 'SAS', 'TOR', 'UTA', 'WAS']}
```

## Create Hierarchical Model

### 2018-2019

```python
# 2018-2019 only
df18 = nba.query("Year =='18-19'")

with pm.Model() as hierarchical_model_1819:
    kappa_minus2 = 82 - pm.Uniform('kappa_minus2', 2, 84)
    kappa = pm.Deterministic('kappa', kappa_minus2 + 2)
    omega = pm.Deterministic('omega', pm.math.dot(1/(kappa - 2), ((df18['Proj']/82) - 1)))
    
    theta = pm.Beta('theta', alpha=omega*(kappa-2)+1, beta=(1-omega)*(kappa-2)+1, shape=n_teams)
            
    y = pm.Binomial('Wins', n=n_teams, p=theta, observed=df18['Wins'])

pm.model_to_graphviz(hierarchical_model_1819)
```

```python
with hierarchical_model_1819:
    trace_1819 = pm.sample(50000, cores=4, nuts_kwargs={'target_accept': 0.95})
```

```python
plt.figure(figsize=(6,14))
pm.forestplot(trace_1819, var_names=['theta']);
```

### 2019-2020

```python
# 2019-2020 only
df19 = nba.query("Year =='19-20'")

with pm.Model() as hierarchical_model_1920:
    kappa_minus2 = 82 - pm.Uniform('kappa_minus2', 2, 84)
    kappa = pm.Deterministic('kappa', kappa_minus2 + 2)
    
    omega = pm.Deterministic('omega', pm.math.dot(1/(kappa - 2), ((df19['Proj']/82) - 1)))
        
    theta = pm.Beta('theta', alpha=omega*(kappa-2)+1, beta=(1-omega)*(kappa-2)+1, shape=n_teams)            

    y = pm.Binomial('Wins', n=n_teams, p=theta, observed=df19['Wins'])
```

```python
with hierarchical_model_1920:
    trace_1920 = pm.sample(50000, cores=4, nuts_kwargs={'target_accept': 0.95})
```

#### Results 2019-2020

```python
pm.summary(trace_1920)
```
